```python
import pandas as pd
import numpy as np

# قراءة ملف البيانات
file_path = '/content/drive/MyDrive/15-6-2024/cumulative2.csv'
df = pd.read_csv(file_path)

# قائمة الأعمدة التي سيتم ملء القيم المفقودة فيها بمتوسط القيم
columns_to_fill = [
    'koi_score', 'koi_period_err1', 'koi_period_err2', 'koi_time0bk_err1', 'koi_time0bk_err2',
    'koi_impact', 'koi_impact_err1', 'koi_impact_err2', 'koi_duration_err1', 'koi_duration_err2',
    'koi_depth', 'koi_depth_err1', 'koi_depth_err2', 'koi_prad', 'koi_prad_err1', 'koi_prad_err2',
    'koi_teq', 'koi_teq_err1', 'koi_teq_err2', 'koi_insol', 'koi_insol_err1', 'koi_insol_err2',
    'koi_model_snr', 'koi_tce_plnt_num', 'koi_steff', 'koi_steff_err1',
    'koi_steff_err2', 'koi_slogg', 'koi_slogg_err1', 'koi_slogg_err2', 'koi_srad', 'koi_srad_err1',
    'koi_srad_err2', 'koi_kepmag'
]

# ملء القيم المفقودة بمتوسط القيم لكل عمود
for column in columns_to_fill:
    if column in df.columns and np.issubdtype(df[column].dtype, np.number):
        mean_value = df[column].mean()
        df[column].fillna(mean_value, inplace=True)

# العثور على القيم المفقودة في العمود 'kepler_name'
missing_indices = df[df['kepler_name'].isna()].index

# إنشاء أرقام متسلسلة للقيم المفقودة في العمود 'kepler_name'
sequence = [f"name-{i+1}" for i in range(len(missing_indices))]

# استبدال القيم المفقودة في العمود 'kepler_name' بالعبارة المتبوعة برقم متسلسل
df.loc[missing_indices, 'kepler_name'] = sequence

# عرض عدد الصفوف والأعمدة بعد التنظيف
num_rows, num_columns = df.shape
print(f"عدد الصفوف: {num_rows}")
print(f"عدد الأعمدة: {num_columns}")

# عرض DataFrame بعد ملء القيم المفقودة
print(df.head())

# إذا أردت حفظ التعديلات، يمكنك حفظ DataFrame كملف CSV
df.to_csv('/content/drive/MyDrive/15-6-2024/cumulative2_cleaned.csv', index=True)
```

```python
import pandas as pd

# قراءة ملف البيانات
file_path = '/content/drive/MyDrive/15-6-2024/cumulative2.csv'
df = pd.read_csv(file_path)

# قائمة الأعمدة النصية التي سيتم ملء القيم المفقودة فيها بالقيمة الأكثر تكرارًا
columns_to_fill = ['koi_teq_err1', 'koi_teq_err2', 'koi_tce_delivname']

# ملء القيم المفقودة بالقيمة الأكثر تكرارًا لكل عمود
for column in columns_to_fill:
    if column in df.columns:
        # تحقق مما إذا كان العمود يحتوي على قيم غير مفقودة
        if df[column].dropna().empty:
            print(f"العمود {column} يحتوي فقط على قيم مفقودة.")
        else:
            mode_value = df[column].mode()[0]
            df[column].fillna(mode_value, inplace=True)

# عرض عدد الصفوف والأعمدة بعد التنظيف
num_rows, num_columns = df.shape
print(f"عدد الصفوف: {num_rows}")
print(f"عدد الأعمدة: {num_columns}")

# عرض DataFrame بعد ملء القيم المفقودة
print(df.head())

# حفظ التعديلات كملف CSV
df.to_csv('/content/drive/MyDrive/15-6-2024/cumulative2_cleaned.csv', index=True)
```

```python
import pandas as pd
import numpy as np

# قراءة ملف البيانات
file_path = '/content/drive/MyDrive/15-6-2024/cumulative2_modified.csv'
df = pd.read_csv(file_path)

# الأعمدة التي سيتم التعامل معها
columns_to_process = [
    'kepler_name', 'koi_score', 'koi_period_err1', 'koi_period_err2', 'koi_time0bk_err1', 'koi_time0bk_err2',
    'koi_impact', 'koi_impact_err1', 'koi_impact_err2', 'koi_duration_err1', 'koi_duration_err2',
    'koi_depth', 'koi_depth_err1', 'koi_depth_err2', 'koi_prad', 'koi_prad_err1', 'koi_prad_err2',
    'koi_teq', 'koi_insol', 'koi_insol_err1', 'koi_insol_err2', 'koi_model_snr', 'koi_tce_plnt_num',
    'koi_tce_delivname', 'koi_steff', 'koi_steff_err1', 'koi_steff_err2', 'koi_slogg', 'koi_slogg_err1',
    'koi_slogg_err2', 'koi_srad', 'koi_srad_err1', 'koi_srad_err2', 'koi_kepmag'
]

# حذف الأعمدة الفارغة
df.dropna(axis=1, how='all', inplace=True)

# ملء القيم الرقمية المفقودة بمتوسط القيم لكل عمود
numeric_columns = df.select_dtypes(include=np.number).columns.tolist()
for column in numeric_columns:
    mean_value = df[column].mean()
    df[column].fillna(mean_value, inplace=True)

# استبدال القيم المفقودة في العمود kepler_name
missing_kepler_name_index = df['kepler_name'].isnull()
for idx in missing_kepler_name_index.index:
    if missing_kepler_name_index[idx]:
        df.loc[idx, 'kepler_name'] = f'name_{idx}'

# عرض عدد الصفوف والأعمدة بعد التنظيف
num_rows, num_columns = df.shape
print(f"عدد الصفوف: {num_rows}")
print(f"عدد الأعمدة: {num_columns}")

# حفظ التعديلات كملف CSV
df.to_csv('/content/drive/MyDrive/15-6-2024/cumulative2_processed.csv', index=True)
```
